```python
def sort_centers(centers, img_path, visualize=False):

    # Define the coordinates of the 6th point
    sixth_point = (40, 190)

    # Calculate distances from the sixth point to each of the centers
    distances = [np.sqrt((x - sixth_point[0])**2 + (y - sixth_point[1])**2) for x, y in centers]

    # Pair each center with its distance from the sixth point
    center_distances = list(zip(centers, distances))

    # Sort the centers by their distance from the sixth point
    sorted_centers = sorted(center_distances, key=lambda x: x[1])

    # Extract the sorted centers and their order
    sorted_centers_only = [center for center, distance in sorted_centers]

    if visualize:
        # Load the provided image
        img = Image.open(img_path)
        # Convert to RGB to plot color on top of the original image
        img_rgb = img.convert('RGB')
        draw = ImageDraw.Draw(img_rgb)

        # Draw the sixth point
        draw.ellipse((sixth_point[0]-3, sixth_point[1]-3, sixth_point[0]+3, sixth_point[1]+3), fill='blue', outline='blue')

        # Draw the centers, lines, and order
        for num, (x, y) in enumerate(sorted_centers_only):
            # Draw the center
            draw.ellipse((x-3, y-3, x+3, y+3), fill='red', outline='red')
            # Draw line from the center to the sixth point
            draw.line((x, y, sixth_point[0], sixth_point[1]), fill='green', width=1)
            # Annotate the order next to the center
            draw.text((x+5, y), f'{num+1}', fill='purple',)

        plt.imshow(img_rgb)
        plt.colorbar()
        plt.show()

    return sorted_centers_only
```

```python
import os
import csv
from collections import defaultdict, Counter

# Путь к папке с файлами (замените на ваш путь)
directory_path = '../data/interim/objects/'

# Словарь для хранения данных о группах и цифрах
groups = defaultdict(lambda: {'numbers': [], 'icons': []})

# Чтение имен файлов и заполнение словаря
for filename in os.listdir(directory_path):
    if filename.endswith('.jpg'):
        parts = filename.split('_')
        group_id = parts[0]
        if 'icon' in parts[1]:
            groups[group_id]['icons'].append(filename)
        else:
            groups[group_id]['numbers'].append(int(parts[1]))

# Функция для определения правильного ответа
def find_correct_answer(numbers):
    count = Counter(numbers)
    for num, freq in count.items():
        if freq > 1:
            return num + 1  # Правильный ответ - повторяющееся число + 1
    return None  # Если не найден повторяющийся элемент

# Создание записей для CSV файла
csv_rows = []
for group_id, data in groups.items():
    correct_answer = find_correct_answer(data['numbers'])
    if correct_answer is None:
        continue  # Пропускаем группы без повторяющихся чисел
    for i in range(2, 7):  # Для иконок от 2 до 6
        row = [
            f'{group_id}_icon_1.jpg',  # Иконка 1
            f'{group_id}_icon_{i}.jpg',  # Иконка i
            'True' if (7 - i) == correct_answer else 'False'  # Правильный/неправильный ответ
        ]
        csv_rows.append(row)

# Запись результатов в CSV файл
csv_file_path = '../data/interim/Clip_dataset.csv'
with open(csv_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Icon 1', 'Icon i', 'Answer'])  # Заголовок
    writer.writerows(csv_rows)

print(f'Done! Results are saved to {csv_file_path}')
```

```python
from matplotlib import pyplot as plt
from collections import defaultdict, Counter
import csv
import json
import os
from PIL import Image

def calculate_center(points):
    # points is a list of two tuples: [(x1, y1), (x2, y2)]
    x_center = (points[0][0] + points[1][0]) / 2
    y_center = (points[0][1] + points[1][1]) / 2
    return (x_center, y_center)
def sort_centers(centers, img_path, visualize=False):

    # Define the coordinates of the 6th point
    sixth_point = (40, 190)

    # Calculate distances from the sixth point to each of the centers
    distances = [np.sqrt((x - sixth_point[0])**2 + (y - sixth_point[1])**2) for x, y in centers]

    # Pair each center with its distance from the sixth point
    center_distances = list(zip(centers, distances))

    # Sort the centers by their distance from the sixth point
    sorted_centers = sorted(center_distances, key=lambda x: x[1])

    # Extract the sorted centers and their order
    sorted_centers_only = [center for center, distance in sorted_centers]

    if visualize:
        # Load the provided image
        img = Image.open(img_path)
        # Convert to RGB to plot color on top of the original image
        img_rgb = img.convert('RGB')
        draw = ImageDraw.Draw(img_rgb)

        # Draw the sixth point
        draw.ellipse((sixth_point[0]-3, sixth_point[1]-3, sixth_point[0]+3, sixth_point[1]+3), fill='blue', outline='blue')

        # Draw the centers, lines, and order
        for num, (x, y) in enumerate(sorted_centers_only):
            # Draw the center
            draw.ellipse((x-3, y-3, x+3, y+3), fill='red', outline='red')
            # Draw line from the center to the sixth point
            draw.line((x, y, sixth_point[0], sixth_point[1]), fill='green', width=1)
            # Annotate the order next to the center
            draw.text((x+5, y), f'{num+1}', fill='purple',)

        plt.imshow(img_rgb)
        plt.colorbar()
        plt.show()

    return sorted_centers_only

# Дополнительно импортируем numpy и PIL для работы с изображениями
import numpy as np
from PIL import ImageDraw, Image

# Предположим, что функция sort_centers уже определена

# Путь к папке с файлами (замените на ваш путь)
directory_path = '../data/interim/objects/'
json_folder_path = '../data/raw/'

# Словарь для хранения данных о группах и цифрах
groups = defaultdict(lambda: {'numbers': [], 'icons': []})

# Чтение имен файлов и заполнение словаря
for filename in os.listdir(directory_path):
    if filename.endswith('.jpg'):
        parts = filename.split('_')
        group_id = parts[0]
        if 'icon' in parts[1]:
            groups[group_id]['icons'].append(filename)
        else:
            groups[group_id]['numbers'].append(int(parts[1]))

# Функция для определения правильного ответа
def find_correct_answer(numbers):
    count = Counter(numbers)
    for num, freq in count.items():
        if freq > 1:
            return num  # Возвращаем повторяющееся число без изменений
    return None  # Если не найден повторяющийся элемент

# Создание записей для CSV файла
csv_rows = []

for group_id, data in groups.items():
    # Загрузка соответствующего JSON-файла
    json_file_path = os.path.join(json_folder_path, f"{group_id}.json")
    with open(json_file_path, 'r') as file:
        json_data = json.load(file)
    
    # Вычисление центров для каждого объекта
    centers = [calculate_center(shape['points']) for shape in json_data['shapes'] if shape['shape_type'] == 'rectangle']
    
    # Использование функции sort_centers для получения отсортированного списка
    sorted_centers_only = sort_centers(centers, None, False)  # Визуализация выключена
    
    # Отображение индексов отсортированных центров на их порядковые номера
    center_order = {center: i+1 for i, center in enumerate(sorted_centers_only)}
    
    correct_answer = find_correct_answer(data['numbers'])
    if correct_answer is None:
        continue  # Пропускаем группы без повторяющихся чисел
        
    for i, center in enumerate(sorted_centers_only):
        icon_number = center_order[center] + 1
        row = [
            f'{group_id}_icon_1.jpg',  # Иконка 1
            f'{group_id}_icon_{icon_number}.jpg',  # Иконка по порядку в sorted_centers_only
            'True' if icon_number == correct_answer else 'False'  # Правильный/неправильный ответ
        ]
        csv_rows.append(row)

# Запись результатов в CSV файл
csv_file_path = '../data/interim/Clip_dataset.csv'
with open(csv_file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Icon 1', 'Icon i', 'Answer'])  # Заголовок
    writer.writerows(csv_rows)

print(f'Done! Results are saved to {csv_file_path}')
```
