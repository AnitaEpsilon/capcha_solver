```python
import easyocr

# Создание объекта reader
reader = easyocr.Reader(['en'], gpu=False)  # Укажите 'gpu=True', если доступна GPU

# Функция для распознавания текста с изображения
def recognize_text_from_image(image_path):
    result = reader.readtext(image_path, detail=1, paragraph=False)
    
    for detection in result:
        bbox, text, confidence = detection
        print(f"Обнаруженный текст: {text}, Уверенность: {confidence*100:.2f}%")
```

```python
from tqdm import tqdm
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def display_image(image_path):
    try:
        img = mpimg.imread(image_path)
        plt.imshow(img)
        plt.axis('off')  # Убрать оси координат
        plt.show()
    except Exception as e:
        print("Ошибка при отображении изображения:", e)
```

```python
# Путь к вашему изображению
image_path = 'images/1.jpg'

# Распознавание текста с изображения
recognize_text_from_image(image_path)
display_image(image_path)
```

```python
for i in tqdm(range(1, 4)):
    image_path = f'images/{i}.jpg'
    recognize_text_from_image(image_path)
    display_image(image_path)
```
